```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("players_data.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df["Goal_Contribution"] = df["Goals"] + df["Assists"]
df
```

```python
top_scorer = df.loc[df["Goals"].idxmax()]
top_scorer
```

```python
position_goals = df.groupby("Position")["Goals"].sum()
position_goals.plot(kind="bar", title="Goals by Position")
plt.ylabel("Goals")
plt.show()
```

```python
plt.scatter(df["Age"], df["Goals"])
plt.title("Age vs Goals")
plt.xlabel("Age")
plt.ylabel("Goals")
plt.show()
```

```python
best_player = df.loc[df["Goal_Contribution"].idxmax()]
best_player
```
